```python
import sys

ID_TOKEN = 0
INT_TOKEN = 1
LEXEME = 2
ERROR = 3
END_OF_INPUT = 4

line = 1; 
[ID_TOKEN, "x"]

def lexInt(input, sign):
    i = 0
    while i < len(input) and input[i].isdigit():
       i = i+1;
    return [[INT_TOKEN, sign*int("".joing(input[0:i]))], input[i:]]

def handleOpeningParenthese(input):
    

def isIdChar(c):
    return c == "_" or c.isalpha() or c.isdigit()
    
def lookup(lexeme):
    # print, get, if, then, else, while, do, end, and, or, not
    if lexeme == "get" or lexeme == "print" or lexeme == "if" or lexeme == "then" or lexeme == "else" or lexeme == "while" or lexeme == "do" or lexeme == "end" or lexeme == "and" or lexeme == "or" or lexeme == "not":
        return [LEXEME, lexeme]
    else:
        return [ID_TOKEN, lexeme]
        

def lexIdOrKeyword(input):
    i = 0
    while i < len(input) and isIdChar(input[i]):
        i = i + 1
    return ["".join(input[0:i])), input[i:]]

# Represents the Start state
def lex(input):
    global line
    # take care of white space
    i = 0
    while i < len(input) and input[i].iswhitespace():
        if input[i] == "\n":
            line = line + 1
        i = i + 1
        
    # check if we are at the end of input
    if i >= len(input):
        return [[END_OF_INPUT, ""], input]
    
    input = input[i:] # remove the whitespace
     
    if input[0] == "=":
        if len(input) > 1 && input[1] == "=":
            return [[LEXEME, "=="], input[2:]]
        else:
            return [[LEXEME, "="], input[1:]]

    elif input[0] == ";":
        # Transition on ';'
        return [[LEXEME, ";"], input[1:]]
    
    elif input[0] == "<":
        if len(input) > 1 && input[1] == "=":
            return [[LEXEME, "<="], input[2:]]
        else:
            return [[LEXEME, "<"], input [1:]]

    elif input[0] == "-":
        if input[1].isdigit():
            return lexInt(input[1:], -1)
        else:
            return [[LEXEME, "-"], input[1:]]

    elif input[0].isdigit():
        return lexInt(input, 1)

    elif input[0] == "+":
        if len(input) > 1 and input[1].isdigit():
            return lexInt(input[1:], 1)
        else:
            return [[ERROR, "Expected digit after '+' on line " + str(line)], input]

    elif input[0] == "*":
        if len(input) > 1 and input[1].isdigit():
            return lexInt(input, 1)
        else:
            return [[ERROR, "Expected digit after '*' on line " + str(line)], input]
            
    elif input[0] == "/":
        if len(input) > 1 and input[1].isdigit():
            return lexInt(input, 1)
        else:
            return [[ERROR, "Expected digit after '/' on line " + str(line)], input]
    
    elif input[0] == "%":
        if len(input) > 1 and input[1].isdigit():
            return lexInt(input, 1)
        else:
            return [[ERROR, "Expected digit after '%' on line " + str(line)], input]

    elif input[0] == "!":
        if len(input) > 1 && input[1] == "=":
            return [[LEXEME, "!="], input[2:]]
            # if len(input) > 2	&& isIdChar(input[2]):
            #   return lexIdOrKeyword(input, 2)
        else:
            return [[ERROR, "Expected '=' after '!' on line " + str(line)], input]

    elif input[0] == "(":
        return handleOpeningParenthese(input)
            
    elif input[0] == "_" || input[0].isalpha():
        return lexIdOrKeyword(input)
    elif input[0].iswhitespace():
        pass
    else: # not a valid lexeme or token
        return[[ERROR, "Unexpected character on line " + str(line)], input]


input = list(sys.stdin.read())
# Loop through the lexemes and tokens until an error or end of input happens
[next, input] = lex(input)

while next[0] != END_OF_INPUT and next[0] != ERROR:
    print(next)
    [next, input] = lex(input)

if next[0] == ERROR:
    print("ERROR: " + next[1])
```
